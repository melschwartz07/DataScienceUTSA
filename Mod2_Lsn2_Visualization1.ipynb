## Plotting a Line Plot ##

There are a variety of features that you have control over for a plot.  They are:

__Line Color__, __Line Width__, __Line Style__, __Line Opacity__, and __Markers__

Line color, line width and style are all included as extra keyword arguments in the plt.plot() function call.

'plt.plot(<x-data>,<y-data>,
    linewidth=<float or int>
    linestyle = '<linestyle abbreviation>',
    color = '<color abbreviation>',
    marker = '<marker abbreviation>')'
    
The linewidth is in pixels, ranging from 0.5 to 3 or more.
    
The common linestyles are:
    
|  Python code    |  Description                 |
|-----------------|------------------------------|
|'-' or 'solid'   |solid line (default)          |
|'--' or 'dashed' |  dashed line                 |
|'-.' or 'dashdot'|  dash-dot line               |
|':' or 'dotted'  | dotted line                  |
| 'None' or ' '   |  No line'                    |
|-----------------|------------------------------|

__Example 5:__
What I don't like about this is that the y-axis starts at 0, which means the changes in costs, which is what we are trying to look at, is buried at the top of the graph.  It would be better if the y axis was scaled.  Let's do that. To do this, you need to import a new library - $math$.  The new line sets the y axis limits to the actual values, with a padding of half (0.5) of the difference.

```python
#Example 5
import math
low = min(MeanByMonth)
high = max(MeanByMonth)
MeanByMonth.plot.bar()
plt.ylim([math.ceil(low-0.5*(high-low)), math.ceil(high+0.5*(high-low))])
plt.xlabel('Month')
plt.ylabel('Cost in Dollars')
```

Now you can really see the differences in costs.

## Subplots ##

Now how do we plot multiple graphs on the same figure?  We use 'subplot'. 
In subplot, we list the number of rows, followed by the number of columns, and then
describe each graph.

__Example 10:__

```python
#Example 10
fig, axes = plt.subplots(2,2)                     # my output will be a 2x2 figure (4 graphs)
axes[0, 0].hist(df['data science'])               # our first histogram!
axes[0, 1].scatter(df['Mes'], df['data science'])  # our first scatterplot
axes[1, 0].plot(df['Mes'], df['machine learning'])
axes[1, 1].plot(df['Mes'], df['deep learning'])
```

Yeah!!!  Fancy!  Lots going on here!

Let's break it down into it's component parts.

## Scatter plot ##

A scatterplot plots 2 variables against each other, with the independent variable along the
x-axis and the dependent on the y-axis.  Here we are plotting the terms 'data science' against the 'machine learning' terms.

__Example 11:__

```python
#Example 11
plt.scatter(df['data science'], df['machine learning'])
plt.xlabel('frequency of data science')
plt.ylabel('frequency of machine learning')
plt.title('Searches for data science vs machine learning')
plt.grid(True)
```

On a scatterplot, if the x and y axis are scaled the same, a perfect 1:1 correlation would be a 45 degree angle.  This graph is scaled the same, so deviations from the 45 degree angle show the higher frequency term choice.  In the lower values, the line goes below 45 degrees, so data science was chosen more.  At about 30 searches, the angle changes to above 45, so machine learning was a more popular choice and then, above about 60, it returns to data science.  

A scatter plot does not have a sense of time, but merely plots 2 values against each other.  We cannot tell which set of search values happened first, second or at any specific time.  You would use one of our line graphs to use time as a parameter.


## Error bar charts ##

Error bar charts are a great way to show the variability in your data.  As a scientist, you can pick either the standard deviation (frequently chosen) or the confidence interval (95 percent).

You first need to calculate the mean and standard deviation of your data.  Let's do that with our search counts.

__Example 12:__

```python
#Example 12
MeanByMonth = df.mean(axis=0)  # calculate the mean of each column
StdByMonth = df.std(axis=0)    # calculate the std of each column

data=[['data science', 20.95,23.95],['machine learning',27.4,28.1],['deep learning',24.2,34.5]]

dfErrorBar = pd.DataFrame(data,columns=['Category','mean','std'])

dfErrorBar
```

Now that we have some numbers, we can anticipate what the graph should look like - mean in the 20s, and large standard deviations.

__Example 13:__

```python
#Example 13
plt.errorbar(dfErrorBar['Category'],dfErrorBar['mean'],yerr=dfErrorBar['std'], fmt = 'o', 
             color='Black', elinewidth=3,capthick=3,errorevery=1, alpha=1, ms=4, capsize = 5)
plt.title('Error Bar of Searches')
plt.grid(True)
```

Looking at this error bar chart, you can tell the data has a lot of variability - is it realistic that you can have negative number of searches - look at the bottom of the error bar chart - they are below zero.
