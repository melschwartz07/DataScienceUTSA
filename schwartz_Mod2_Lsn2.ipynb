```python
# Setup
import matplotlib.pyplot as plt  # for generating plots
import matplotlib
import pandas as pd     # import the pandas library to read a .csv file
import numpy as np      # use py for numpy from now on
# this command tells Jupyter Notebook to plot in this space
%matplotlib inline

median_home_values = pd.read_csv("Median_List_raw_month.csv")   # make sure the data file is in the same directory

MeanByMonth = median_home_values.mean(axis=0)  # calculate the mean of each column
MedianByMonth = median_home_values.median(axis=0)
```

```python
#Example 1
plt.plot(MeanByMonth)   # generates the plot
plt.show()              # prints the plot here
```

```python
#Example 2
labels = ['12/2017','6/2018','12/2018','6/2019','1/2020']  # the new x tick marks
x = np.arange(0,35,7)                       # the label locations
fig,ax = plt.subplots()
ax.set_xticks(x)
ax.set_xticklabels(labels)                  # all of the above was for the fancy x tick labels
plt.plot(MeanByMonth)
plt.title('Median Housing Cost by Month')    # title
plt.xlabel('Month')                          # xlabel
plt.ylabel('Cost in dollars')                # ylabel
plt.show()                                   # must be last item
```

```python
#Example 3
fig,ax = plt.subplots()             # New for multi line graphs set up parameters

labels = ['12/2017','6/2018','12/2018','6/2019','1/2020']  # the new x tick marks
x = np.arange(0,35,7)                       # the label locations
ax.set_xticks(x)
ax.set_xticklabels(labels)                  # all of the above was for the fancy x tick labels

# new for multi line graphs
ax.plot(MeanByMonth)
ax.plot(MedianByMonth)

ax.set_title('Median Purchase Cost of Housing in 35 US Regions')
ax.legend(['Mean','Median'])
plt.ylabel('Cost in dollars')                # ylabel

plt.show()
```

Question 1: Which of these, line graph (using on the median line) or the bar chart is easier to read? There are times when you would use either one - when youre trying to focus on a specific idea - Why would you use a line graph for this data? Why would you use a bar chart for this data?

The line graph was easier to read because it is easier for the viewer to spot a trend over a 6 month period, rather than a much more condensed month to month change as we saw in the bar chart. As stated in the Visualization document, "graphs are used to display relationships among multiple quantitative values by giving them shape". We see a much more fluid shape with a graph because the changes are much more continuous, whereas a bar chart appears more "clunky". A bar chart would be useful for this data if the trends were being shown over a longer period of time. The median housing cost would have more impact on the viewer if we were viewing a year to year change.

```python
#Example 5
import math
low = min(MeanByMonth)
high = max(MeanByMonth)
MeanByMonth.plot.bar()
plt.ylim([math.ceil(low-0.5*(high-low)), math.ceil(high+0.5*(high-low))])
plt.xlabel('Month')
plt.ylabel('Cost in Dollars')
```
